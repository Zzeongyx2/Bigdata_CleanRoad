```python
import pandas as pd
```

### 데이터 불러오기

```python
data1 = pd.read_csv('data/subway1.csv', encoding='CP949')
data2 = pd.read_csv('data/subway2.csv', encoding='CP949')
data3 = pd.read_csv('data/subway3.csv', encoding='CP949')
data4 = pd.read_csv('data/subway4.csv', encoding='CP949')
data5 = pd.read_csv('data/subway5.csv', encoding='CP949')
data6 = pd.read_csv('data/subway6.csv', encoding='CP949')
data7 = pd.read_csv('data/subway7.csv', encoding='CP949')
data8 = pd.read_csv('data/subway8.csv', encoding='CP949')
data9 = pd.read_csv('data/subway9.csv', encoding='CP949')
data10 = pd.read_csv('data/subway10.csv', encoding='CP949')
data11 = pd.read_csv('data/subway11.csv', encoding='CP949')
data12 = pd.read_csv('data/subway12.csv', encoding='CP949')
```

### 데이터 병합

```python
data = pd.concat([data1, data2, data3, data4, data5, data6,
                  data7, data8, data9, data10, data11, data12])
```

```python
data
```

### 일자별 역별 승하차 총 승객수

```python
subway = data.groupby(['사용일자','역명'])['승차총승객수','하차총승객수'].sum()
```

```python
subway
```

```python
subway['승하차총승객수'] = subway['승차총승객수'] + subway['하차총승객수']
```

```python
subway
```

```python
sum_p = subway.drop(['승차총승객수', '하차총승객수'], axis=1)
```

```python
sum_p
```

### 일자별 역별 승하차 평균 승객수

```python
avg_p = subway.groupby('역명')['승차총승객수','하차총승객수'].mean()
```

```python
avg_p
```

```python
avg_p['승하차평균승객수'] = (avg_p['승차총승객수'] + avg_p['하차총승객수']) / 2
```

```python
avg_p
```

```python
avg_p = avg_p.drop(['승차총승객수', '하차총승객수'], axis=1)
```
