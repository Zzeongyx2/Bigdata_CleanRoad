```python
import pandas as pd
```

### 데이터 불러오기

```python
data1 = pd.read_csv('data/미세먼지_2017.csv', parse_dates=["측정일자"], encoding='CP949')
data2 = pd.read_csv('data/미세먼지_2018.csv', parse_dates=["측정일자"], encoding='CP949')
data3 = pd.read_csv('data/미세먼지_2019.csv', parse_dates=["측정일자"], encoding='CP949')
data4 = pd.read_csv('data/미세먼지_2020.csv', parse_dates=["측정일자"], encoding='CP949')
```

```python
tem = pd.read_csv('data/평균기온.csv', parse_dates=["일시"], encoding='CP949')
```

```python
high_tem = pd.read_csv('data/구별일자별최고기온.csv', parse_dates=["측정일자"], encoding='CP949')
```

```python
spot = pd.read_csv('data/관측지점.csv', encoding='CP949')
```

### 미세먼지 데이터 병합

```python
air = pd.concat([data1,data2,data3,data4])
```

### 데이터 확인

```python
air
```

```python
tem
```

```python
spot
```

### 관측소별 일별 평균기온 

```python
avg_tem = tem.groupby('지점명').mean()
```

```python
avg_tem
```

### 관측소 좌표 X 평균기온

```python
spot_avg_tem = pd.merge(spot, avg_tem)
```

```python
spot_avg_tem
```

### 자치구별 일별 미세먼지 평균

### 구별 폭염일수

```python
hotday = high_tem[high_tem['최고기온'] >= 33]
```

```python
hotday
```

```python
hotday = hotday.groupby('지점명').count()
```

```python
hotday = hotday.drop(['측정일자'], axis=1)
hotday.columns = ['폭염일수']
```
